```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay as cmdis
import seaborn as sns
import numpy as np
import pandas as pd
import os

%matplotlib inline
```

```python
cm_ECG = np.array([[1070, 224], [98, 1140]])
cm_SpO2 = np.array([[279, 16], [9, 283]])
labels = np.array(["No OSA", "OSA"])
```

```python
cm_dis = cmdis(confusion_matrix=cm_ECG, display_labels=["No OSA", "OSA"])
cm_dis.plot()
```

```python
cm_dis = cmdis(confusion_matrix=cm_SpO2, display_labels=["No OSA", "OSA"])
cm_dis.plot()
```

```python
output_file_path = os.path.join(os.getcwd(), 'confusion_matrix_ECG.png')
confusion_df = pd.DataFrame(cm_ECG, index=labels, columns=labels)
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.5)

sns.heatmap(confusion_df, annot=True, fmt='d', cmap="Blues", cbar_kws={'label': 'Count'})

plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for ECG')

plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.savefig(output_file_path, dpi=300)
plt.show()
```

```python
output_file_path = os.path.join(os.getcwd(), 'confusion_matrix_SpO2.png')
confusion_df = pd.DataFrame(cm_SpO2, index=labels, columns=labels)
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.5)

sns.heatmap(confusion_df, annot=True, fmt='d', cmap="Blues", cbar_kws={'label': 'Count'})

plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for SpO2')

plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.savefig(output_file_path, dpi=300)
plt.show()
```
