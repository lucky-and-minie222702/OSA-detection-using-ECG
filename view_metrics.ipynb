```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay as cmdis
import numpy as np

%matplotlib inline
```

```python
cm_ECG = np.array([[1070, 224], [98, 1140]])
cm_SpO2 = np.array([[279, 16], [9, 283]])
```

```python
cm_dis = cmdis(confusion_matrix=cm_ECG, display_labels=["No OSA", "OSA"])
cm_dis.plot()
```

```python
cm_dis = cmdis(confusion_matrix=cm_SpO2, display_labels=["No OSA", "OSA"])
cm_dis.plot()
```
