```python
import numpy as np
import matplotlib.pyplot as plt
from os import path

%matplotlib inline
```

```python
loss = np.load("history/SpO2_generate_loss.npy")
mae = np.load("history/SpO2_generate_mae.npy")
mse = np.load("history/SpO2_generate_mse.npy")
```

```python
plt.figure(figsize=(12, 7))
plt.title("Training history")
plt.plot(loss, label="Huber loss")
plt.plot(mae, label="MAE")
plt.plot(mse, label="MSE")
plt.legend()
plt.yticks([x / 200 for x in range(20)])
plt.xlabel("Epochs")
plt.show()
```
